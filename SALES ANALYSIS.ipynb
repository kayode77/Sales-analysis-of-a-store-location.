# Sales types in Astonia in relation to the product type, the total bill and the month of Purchase

```python
Size_regular=df.query('size == "Regular"')
Size_Large=df.query('size == "Large"')
Size_Not_Defined=df.query('size == "Not Defined"')
Size_Small =df.query('size == "Small"')
```

```python
def Analysis_of_store_location_product_type_month_in_relation_to_total_bill_and_size(store_location,products,month_name,Size):
    store_location =df['store_location'] == store_location
    producttype = df['product_type'].str.contains(products)
    month_name=df['month_name'] == month_name
    Total_bill =  Size.loc[store_location &  producttype & month_name, ['store_location','product_type','total_bill']]
    Total_bill_of_product_in_specific_month =Total_bill.pivot_table(values='total_bill',index='product_type', aggfunc='sum')
    return Total_bill_of_product_in_specific_month
```

## Regular sales

```python
Total_bill_of_tea_in_June=Analysis_of_store_location_product_type_month_in_relation_to_total_bill_and_size('Astoria','tea','June',Size_regular)
Total_bill_of_tea_in_June
```

```python
Total_bill_of_tea_in_June.plot(kind='bar',xlabel='Product type', ylabel='Total bill', title='Total bill of tea for the monthy of June on Regular sales', figsize=(8,6))
```

```python
Total_bill_of_tea_in_June.plot(kind='pie',y='total_bill')
```

# Sales types in Hell's Kitchen  in relation to the product type, the total bill,the month of Purchase and time of Purchase

## Time = 10th hour of the day

```python
def Analysis_of_store_location_product_type_month_in_relation_to_total_bill_and_time(store_location,products,month_name,hour):
    store_location =df['store_location'] == store_location
    producttype = df['product_type'].str.contains(products)
    month_name=df['month_name'] == month_name
    Time_of_purchase =df['hour'] == hour
    Total_bill = df.loc[store_location &  producttype & month_name & Time_of_purchase, ['product_type','total_bill']]
    Total_bill_of_product_in_specific_month =Total_bill.pivot_table(values='total_bill',index='product_type', aggfunc='sum')
    return Total_bill_of_product_in_specific_month
```

```python
Total_bill_of_beans_at_tenth_hour = Analysis_of_store_location_product_type_month_in_relation_to_total_bill_and_time("Hell's Kitchen","eans","June",10)
Total_bill_of_beans_at_tenth_hour
```

```python
Total_bill_of_beans_at_tenth_hour.plot(kind='bar',xlabel='Product type', ylabel='Total bill', title="Total bill of beans at tenth hour in June in Hell's Kitchen", figsize=(8,6))
```

```python
Total_bill_of_beans_at_tenth_hour.plot(kind='pie',y='total_bill')
```
